```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv("all_results_10-20-2024.csv")

# Define metrics
metrics = ["metrics.iou", "metrics.f1score", "metrics.precision", "metrics.recall"]

# Get unique architectures
architectures = df["model.model_architecture"].unique()

# Create subplots - one for each architecture
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

# Plot each metric
for idx, metric in enumerate(metrics):
    ax = axes[idx]
    
    # Prepare data for plotting
    plot_data = df[["model.model_architecture", "model.loss", metric]].copy()
    
    # Create grouped box plot
    sns.boxplot(
        data=plot_data,
        x="model.model_architecture",
        y=metric,
        hue="model.loss",
        ax=ax,
        palette="Set2"
    )
    
    # Formatting
    ax.set_title(f"{metric.split('.')[-1].upper()} by Architecture and Loss Function", 
                 fontsize=12, fontweight='bold')
    ax.set_xlabel("Model Architecture", fontsize=10)
    ax.set_ylabel(metric.split('.')[-1].capitalize(), fontsize=10)
    ax.legend(title="Loss Function", loc='best', fontsize=8)
    ax.tick_params(axis='x', rotation=45)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('loss_effect_by_architecture.png', dpi=300, bbox_inches='tight')
plt.show()

# Alternative: Create separate figure for each architecture
for arch in architectures:
    arch_data = df[df["model.model_architecture"] == arch]
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()
    
    for idx, metric in enumerate(metrics):
        ax = axes[idx]
        
        sns.boxplot(
            data=arch_data,
            x="model.loss",
            y=metric,
            ax=ax,
            palette="Set3"
        )
        
        ax.set_title(f"{metric.split('.')[-1].upper()}", fontsize=12, fontweight='bold')
        ax.set_xlabel("Loss Function", fontsize=10)
        ax.set_ylabel(metric.split('.')[-1].capitalize(), fontsize=10)
        ax.tick_params(axis='x', rotation=45)
        ax.grid(axis='y', alpha=0.3)
    
    fig.suptitle(f"Performance Metrics for {arch}", fontsize=14, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.savefig(f'loss_effect_{arch.replace(" ", "_")}.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Load your results file ---
df = pd.read_csv("all_results_10-20-2024.csv")

# --- Select relevant columns ---
metrics = [
    "metrics.iou",
    "metrics.f1score",
    "metrics.precision",
    "metrics.recall"
]

cols = ["model.model_architecture", "model.loss"] + metrics
df_sub = df[cols].copy()

# --- Melt into long form for seaborn ---
melted = df_sub.melt(
    id_vars=["model.model_architecture", "model.loss"],
    value_vars=metrics,
    var_name="Metric",
    value_name="Score"
)

# Clean up metric names for readability
melted["Metric"] = melted["Metric"].str.replace("metrics.", "", regex=False)

# --- Plot ---
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=melted,
    x="model.loss",
    y="Score",
    hue="model.model_architecture",
    col="Metric",
    kind="bar",
    palette="Set2",
    height=4,
    aspect=1.1
)

# --- Style tweaks ---
g.set_titles("{col_name}")
g.set_axis_labels("Loss Function", "Score (01)")

for ax in g.axes.flat:
    ax.set_ylim(0, 1)
    ax.tick_params(axis='x', rotation=30)

plt.subplots_adjust(top=0.85)
plt.suptitle("Effect of Loss Functions on Metrics per Model Architecture", fontsize=14)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Load ---
df = pd.read_csv("all_results_10-20-2024.csv")

# --- Select metrics ---
metrics = ["metrics.iou", "metrics.f1score", "metrics.precision", "metrics.recall"]
metrics = [m for m in metrics if m in df.columns]

# --- Compute max per model/loss combination ---
group_cols = ["model.model_architecture", "model.loss"]
df_max = (
    df[group_cols + metrics]
    .groupby(group_cols, as_index=False)
    .max(numeric_only=True)
)

# --- Melt to long form ---
melted = df_max.melt(
    id_vars=group_cols,
    value_vars=metrics,
    var_name="Metric",
    value_name="Score"
)

melted["Metric"] = melted["Metric"].str.replace("metrics.", "", regex=False)

# --- Plot ---
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, len(metrics), figsize=(4 * len(metrics), 5), sharey=True)
palette = sns.color_palette("Set2")

# --- Plot each metric subplot ---
for i, metric in enumerate(melted["Metric"].unique()):
    ax = axes[i]
    subset = melted[melted["Metric"] == metric]
    sns.barplot(
        data=subset,
        x="model.loss",
        y="Score",
        hue="model.model_architecture",
        palette=palette,
        ax=ax,
        errorbar=None
    )
    ax.set_title(metric)
    ax.set_xlabel("Loss Function")
    if i == 0:
        ax.set_ylabel("Score (01)")
    else:
        ax.set_ylabel("")
    ax.set_ylim(0, 1)
    ax.tick_params(axis="x", rotation=30)

    # Remove individual legends
    ax.get_legend().remove()

# --- Shared legend ---
handles, labels = ax.get_legend_handles_labels()
fig.legend(
    handles,
    labels,
    loc="center left",
    bbox_to_anchor=(1.02, 0.5),
    title="Model Architecture",
    frameon=True
)

fig.suptitle("Maximum Metric Scores per Model Architecture and Loss Function", fontsize=14)
fig.tight_layout(rect=[0, 0, 0.88, 0.93])  # leave space for legend on the right
plt.show()
```
